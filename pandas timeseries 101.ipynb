# Holidays
CustomBusinessDay can be used as frequency in date_range
As we can see 2023-07-04 ( TUE) is not included in the date range
The below is US holiday , pandas_market_calendars can be u

```python
from pandas.tseries.holiday import  USFederalHolidayCalendar
from pandas.tseries.offsets  import CustomBusinessDay
usb = CustomBusinessDay(calendar = USFederalHolidayCalendar())
pd.date_range(start ='2023-07-01' , end='2023-07-09' , freq =usb)
```

## The below is a custom calendar  , 2/1 and 2/10 are holidays , observe on nearest working day
2024-02-10 (Sat) falls on 2024-02-09 (Fri)  , 2024-02-01 is Thur , , final result excludes 2024-02-01 and 2024-02-09

```python
from pandas.tseries.holiday import AbstractHolidayCalendar , Holiday , nearest_workday
class MyHolidayCalendar(AbstractHolidayCalendar):
    """
    US Federal Government Holiday Calendar based on rules specified by:
    https://www.opm.gov/policy-data-oversight/pay-leave/federal-holidays/
    """

    rules = [
        Holiday("Holiday 1", month=2, day=1, observance=nearest_workday),
        Holiday("Holiday 2", month=2, day=10, observance=nearest_workday),
    ]
```

```python

myc = CustomBusinessDay(calendar = MyHolidayCalendar())
pd.date_range(start='2024-02-01' , end='2024-02-10' , freq = myc) 
```

```python
## Egypt has weekend on Friday and Saturday
 weekmask helps in this case 'Sun Mon Tue Wed Thu'
```

```python
myc = CustomBusinessDay( weekmask='Sun Mon Tue Wed Thu')
pd.date_range(start='2024-02-01' , end='2024-02-10' , freq = myc) 
```

## Period
Unlike Timestamp which describes an exact time point Period describes a time span. For example 2023 Q4 describes the window between 2023-10-01 and 2023-12-31

Examples are Year , Quarter , Month , Day , Hour , minute , second ...

```python
year = pd.Period('2023')
print(" {} , {} ".format(year.start_time , year.end_time))
```

## Irregular fiscal year

Most of the companies fiscal year follow calendar year that ends on Dec 31st.  Some of them do not follow this rule. For example Walmart fiscal year ends on Jan 31st
- Q1 Feb 1st to April 30th
- Q2 May 1st to July 31st
- Q3 Aug 1st to Oct 31st
- Q4 Nov 1st to Jan 31st

```python
# Q-JAN indicates period ends in Jan
walmart_q1 = pd.Period('2024Q1' , freq='Q-JAN')
print("  {} , {} ".format( walmart_q1.start_time , walmart_q1.end_time))
```

```python
walmart_q1 = pd.Period('2024Q2' , freq='Q-JAN')
print("  {} , {} ".format( walmart_q1.start_time , walmart_q1.end_time))
```

## MultiIndex

```python
df = pd.read_csv('multiindex_data.csv' , parse_dates=['Date'] ,  dtype =  {'UPC EAN':str})
df
```

```python
# set index
df1 = df.set_index(['Date' , 'Store' , 'Category' , 'Subcategory' , 'Description'])
df1
```

```python
# Append UPC to index as well
df1.set_index('UPC EAN' , append=True , inplace=True)
df1
```

```python
# We want to change the order of the index to move UPC ENA in front of Description
df1.reorder_levels(order=['Date', 'Store', 'Category', 'Subcategory', 'UPC EAN', 'Description'])
```

```python
# Or use swap level
df1.swaplevel('Description', 'UPC EAN')
```

```python
df1.sort_index(inplace=True)
df1
```

```python
df1.loc[('2018-07-12' , 'Store 1' , 'Beer') , :]
```

```python
# Find all 
df1.xs('Store 1' , level='Store')
```

```python
# Find all Beers , keep categtory information
df1.xs('Beer' , level='Category' , drop_level=False)
```

## xs can query multi level index

```python
# Find 2018-07-10 , Beer
df1.xs(('2018-07-10' , 'Beer') , level = ('Date' , 'Category') , drop_level=False)  
```

## Slice

slice does the trick , can be None , can be a range

```python
# To find all Beers  on 2018-07-10, sloce(None) makes the trick , slice
df1.loc[('2018-07-10' , slice(None) , 'Beer') , :]
```

```python
# Find all Beer between 2018-07-10 and 2018-07-11
df1.loc[(slice('2018-07-10' , '2018-07-11') , slice(None) , 'Beer') , :]
```
