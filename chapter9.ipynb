```python
url = 'https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/examples/spx.csv'
from datetime import datetime
fig, ax = plt.subplots()
data = pd.read_csv(url, index_col=0, parse_dates=True)
spx = data["SPX"]

spx.plot(ax=ax, color="black")
crisis_data = [
    (datetime(2007, 10, 11), "Peak of bull market"),
    (datetime(2008, 3, 12), "Bear Stearns Fails"),
    (datetime(2008, 9, 15), "Lehman Bankruptcy")
]
for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor="black", headwidth=4, width=2,
                                headlength=4),
                horizontalalignment="left", verticalalignment="top")
# Zoom in on 2007-2010
ax.set_xlim(["1/1/2007", "1/1/2011"])
ax.set_ylim([600, 1800])

ax.set_title("Important dates in the 2008-2009 financial crisis")
```

### add_patch 

```python
fig, ax = plt.subplots()

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color="black", alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color="blue", alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                   color="green", alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
url = 'https://raw.githubusercontent.com/wesm/pydata-book/3rd-edition/examples/tips.csv'
tips = pd.read_csv(url)
tips
```

```python
party_counts =pd.crosstab(tips['day'] , tips['size'])
party_counts 
```

```python
### drop size == 1 and size == 6
party_count = party_counts.loc[: , [2 , 3 , 4,  5]]
```

```python
party_count.reindex(index=["Thur", "Fri", "Sat", "Sun"])
```

```python
party_pcts  = party_counts.div(party_counts.sum(axis="columns"), axis="index")
party_pcts
```

```python
 party_pcts.plot.bar(stacked=True)
```

### With data that requires aggregation or summarization before making a plot, using the seaborn package can make things much simpler

```python
import seaborn as sns
tips["tip_pct"] = tips["tip"] / (tips["total_bill"] - tips["tip"])
tips.head()
```

```python
### sns plots the mean of tip_pct for the days
### The black lines drawn on the bars represent the 95% confidence interval
```

```python

sns.barplot(x="tip_pct", y="day", data=tips, orient="h")
```

### looks like smoker may pay more tips
hue option that enables us to split by an additional categorical value 

```python
sns.barplot(x="tip_pct", y="day", hue="smoker", data=tips, orient="h")
```

```python
sns.catplot(x="day", y="tip_pct", hue="time", col="smoker",    kind="bar", data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x="day", y="tip_pct", row="time",
           col="smoker",
              kind="bar", data=tips[tips.tip_pct < 1])
```
